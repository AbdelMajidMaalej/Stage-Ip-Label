```python
env = Agent(loaded_dic,loaded_tabs)
```

```python
def update_Q_sarsamin(alpha, gamma, Q, state, action, reward, next_state=None):
    print(f"Next State: {next_state}")
    if next_state == -1:
        return 0
    else:
        """Returns updated Q-value for the most recent experience."""
        current = Q[str(state)][action]  # estimate in Q-table (for current state, action pair)
        Qsa_next = np.min(Q[str(next_state)])  # value of next state 
        target = reward + (gamma * Qsa_next)               # construct TD target
        print(f"Target: {target}")
        new_value = current + (alpha * (target - current)) # get updated value 
        print(f"New Value: {new_value}")
        return new_value
```

```python
def epsilon_greedy(Q, state, nA, eps, action_space):
    """Selects epsilon-greedy action for supplied state.
    
    Params
    ======
        Q (dictionary): action-value function
        state (int): current state
        nA (int): number actions in the environment
        eps (float): epsilonrandom.choice(np.arange(nA))
    """
    if random.random() > eps: # select greedy action with probability epsilon
        return np.argmin(Q[state])
    else:                     # otherwise, select an action randomly
        return random.choice(np.arange(action_space))
```

```python
class NumpyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        return super(NumpyEncoder, self).default(obj)
```

```python
def q_learning(env, num_episodes, alpha, loaded_dic, gamma=1.0, plot_every=1, Q=None, scenarios=None):
    """Q-Learning - TD Control
    
    Params
    ======
        num_episodes (int): number of episodes to run the algorithm
        alpha (float): learning rate
        gamma (float): discount factor
        plot_every (int): number of episodes to use when calculating average score
    """
    nA = max(loaded_dic.values())
    if Q == None:
        Q = {
        key : np.zeros(value)  for key, value in loaded_dic.items()
        } # initialize empty dictionary of arrays
    if scenarios ==None:
        scenarios = [{"score" : 100,"scenario" : []} for i in range(5)]
    # monitor performance
    tmp_scores = deque(maxlen=plot_every)     # deque for keeping track of scores
    avg_scores = deque(maxlen=num_episodes)   # average scores over every plot_every episodes
    
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode%2==0:
            # Save the Q Learning Table using json
            Q_list = {key : array.tolist()  for key,array in Q.items()}
            with open("Q.json", "w") as file:
                json.dump(Q_list, file)
            with open("scenarios.json", "w") as file:
                json.dump({"all_scenarios" : scenarios}, file, cls=NumpyEncoder)
          
        print("\rEpisode {}/{}".format(i_episode, num_episodes), end="\n")
        sys.stdout.flush()
        score = 0                                              # initialize score
        state = env.reset()                                    # start episode
        eps = 40                                  # set value of epsilon
        
        while True:
            print(f"State :{state}")
            action_space = loaded_dic[str(state)] -1
            print(f"Action :{action_space}")
            action = epsilon_greedy(Q, state, nA, eps, action_space)         # epsilon-greedy action selection
            next_state, reward, done, info = env.step(action,state)  # take action A, observe R, S'
            score += reward                                    # add reward to agent's score
            Q[str(state)][action] = update_Q_sarsamin(alpha, gamma, Q, \
                                                 state, action, reward, next_state)        
            state = next_state                                 # S <- S'
            if done:
                print(info)
                print(score)
                tmp_scores.append(score)                       # append score
                #Add the scenario to the info
                if score!=0 and score < scenarios[0]["score"]:
                    scenarios[0] = {"score" : score,"scenario": info["scenario"]}
                    scenarios = sorted(scenarios, key=lambda x: x["score"], reverse=True)
                break
        if (i_episode % plot_every == 0):
            avg_scores.append(np.mean(tmp_scores))
            
    # plot performance
    plt.plot(np.linspace(0,num_episodes,len(avg_scores),endpoint=False), np.asarray(avg_scores))
    plt.xlabel('Episode Number')
    plt.ylabel('Average Reward (Over Next %d Episodes)' % plot_every)
    plt.show()
    # print best episode performance
    print(('Best Average Reward over %d Episodes: ' % plot_every), np.min(avg_scores))
    return Q,scenarios
```

```python
# obtain the estimated optimal policy and corresponding action-value function
Q_sarsamin,scenarios = q_learning(env, 100, .01, loaded_dic)

print(Q_sarsamin)
```
